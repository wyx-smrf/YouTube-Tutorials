# Data Cleaning

---

```python
mask = ((tree_census_subset['status'] == 'Stump') | (tree_census_subset['status'] == 'Dead'))
```

```python
tree_census_subset.loc[mask]
```

```python
tree_census_subset.loc[mask] = tree_census_subset.loc[mask].fillna('Not Applicable')
```

```python
tree_census_subset.loc[mask, 'health']
```

```python
tree_census_subset[tree_census_subset['status'] == 'Stump']
```

```python
tree_census_subset.isna().sum()
```

```python
# Locate the null value in health column

tree_census_subset[tree_census_subset['health'].isna()]
```

```python
tree_census_subset.isna().sum()
```

```python
# For loop

cols = ['problems', 'health', 'spc_latin', 'sidewalk']
replacements = ['None', 'Good', 'No Observation', 'No damage']

for coln, vals in zip(cols, replacements):
    tree_census_subset[coln].fillna(vals, inplace=True)
```

```python
tree_census_subset.isna().sum()
```

```python
big_trees = tree_census_subset[(tree_census_subset['tree_dbh']>60) | (tree_census_subset['stump_diam']>60)]
big_trees
```

```python
tree_census_subst = tree_census_subset[(tree_census_subset['tree_dbh']<=60) | (tree_census_subset['stump_diam']<=60)]
tree_census_subst
```

```python
tcs_alive = tree_census_subst[tree_census_subst['status']=='Alive']
tcs_dead_or_stump = tree_census_subst[(tree_census_subst['status']=='Alive') | (tree_census_subset['status']=='Stump')]
```

```python
# tcs_alive.groupby('spc_latin').mean()

stats_alive = tcs_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','75%']]
stats_alive
```

```python
tcs_alive = tcs_alive.merge(stats_alive, on='spc_latin', how='left')
```

```python
mask = tcs_alive['tree_dbh']<tcs_alive['25%']
tcs_alive.loc[mask,'tree_dbh'] = tcs_alive['25%']

mask = tcs_alive['tree_dbh']>tcs_alive['75%']
tcs_alive.loc[mask,'tree_dbh'] = tcs_alive['75%']
```
